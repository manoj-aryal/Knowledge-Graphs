```python
# You need to write down your own code here
# Task: Given any head entity name (e.g. Q30) and relation name (e.g. P36), you need to output the top 10 closest tail entity names.
# File entity2vec.vec and relation2vec.vec are 50-dimensional entity and relation embeddings.
# If you use the embeddings learned from Problem 1, you will get extra credits.

import numpy as np
import tensorflow as tf
import os
import time
import datetime

def load_file(fname, skip=False):
    file = {}
    with open(fname) as i:
        if skip:
            next(i)
        for line in i:
            (key, val) = line.split()
            file[key] = int(val)
    return file


file_e = load_file('../Problem1/data/entity2id.txt', True)
file_r = load_file('../Problem1/data/relation2id.txt', True)

file_vec_e = np.loadtxt('../Problem1/entity2vec_R.txt')
file_vec_r = np.loadtxt('../Problem1/relation2vec_R.txt')
```

```python
def return_vector(name, file_id, array):
    id = file_id[name]
    vector = array[id]
    return vector

def e_function(h, r, t):
    e = np.linalg.norm(h+r-t, axis=0)
    return e

print('Question No.1:')
print('')
h = return_vector('Q30', file_e, file_vec_e)
r = return_vector('P36', file_r, file_vec_r)
score = [e_function(h,r,t) for t in file_vec_e]



id_sorted = np.argsort(score, axis=0)[0:15]
  
for id in id_sorted:
    for k, v in file_e.items():
        if id == v:
            print(k, v, score[id])
            

print('')
print('Question No.2:')
print('')
h = return_vector('Q30', file_e, file_vec_e)
t = return_vector('Q49', file_e, file_vec_e)
score = [e_function(h,r,t) for r in file_vec_r]



id_sorted = np.argsort(score, axis=0)[0:15]
  
for id in id_sorted:
    for k, v in file_r.items():
        if id == v:
            print(k, v, score[id])
```
